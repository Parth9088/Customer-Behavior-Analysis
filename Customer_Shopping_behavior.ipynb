```python
# Checking NaN values in Clounms, and doing there sum with sum() method.
df.isnull().sum()
```

```python
# we are replacing the NaN values with the median, and we take the median category wise for better Results.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# Now let's convert the Columns names into lower case or we can sayt in "Snake Case". 

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')

# Rename a particular column
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group

labels = ('Young-Adults','Adult','Middle-Aged','Senior')
df['age_group'] = pd.qcut(df['age'], q = 4 , labels = labels)
```

```python
df[['age','age_group']].head(50)
```

```python
# Create new columns purchase_frequency_days

frequency_mapping ={
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Annually' : 365,
    'Bi-Weekly' : 14,
    'Every 3 Month' : 90
}
# Converting the text info into the numerical to make it easy to analys=ze and compare with other.
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(50)
```

```python
# Now let's Check the discount and promo code columns 
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis = 1 )
```
